# Base Performance (Include all raw data)

#### Prepare LabelEncoder and OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

y = df.iloc[:, -1].values
x = df.iloc[:, :-1]

cat_col = [col for col in x.columns if df[col].dtype == 'O']

class get_encoder():
    def __init__(self, cat_col, df):
        self.cat_col = cat_col
        self.df = df

    def fit_transform(self):

        lb_encode = []
        lb_encoder = LabelEncoder()
        oh_encoder = OneHotEncoder(sparse_output=False)

        df_cat = self.df[self.cat_col]

        for col in self.cat_col:
                df_cat[col] = lb_encoder.fit_transform(df_cat[col])
                lb_encode.append(lb_encoder.classes_)

        oh_encoder.fit(df_cat)
            
        return lb_encode, oh_encoder
    
lb_encode, oh_encoder = get_encoder(cat_col, x).fit_transform()
```

#### Clean up missing numerical values

```python
x_train = df_train.iloc[:, :-1]
y_train = df_train.iloc[:, -1].values

cat_col = x_train.dtypes[x_train.dtypes == 'O'].index
num_col = x_train.dtypes[x_train.dtypes != 'O'].index

x_train[num_col].isna().sum()[x_train[num_col].isna().sum() != 0]
x_train['LotFrontage'].fillna(0, inplace=True)
x_train['MasVnrArea'].fillna(0, inplace=True)
x_train['GarageYrBlt'].fillna(0, inplace=True)
```

```python
x_train.head()
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler

class train_preprocessing():
    def __init__(self, cat_col, num_col, df, lb_encode, oh_encoder):
        self.cat_col = cat_col
        self.num_col = num_col
        self.df = df
        self.lb_encode = lb_encode
        self.lb_encoder = LabelEncoder()
        self.s_scaler = StandardScaler()
        self.oh_encoder = oh_encoder
        

    def fit_transform(self):

        # Categorical
        df_cat = self.df[self.cat_col]
        for idx, col in enumerate(self.cat_col):
            self.lb_encoder.fit(self.lb_encode[idx])
            df_cat[col] = self.lb_encoder.transform(df_cat[col])
        
        array_cat = self.oh_encoder.transform(df_cat)
        df_cat = pd.DataFrame(array_cat, columns=self.oh_encoder.get_feature_names_out())

        # Numerical
        df_num = self.df[self.num_col]
        array_num = self.s_scaler.fit_transform(df_num)
        df_num = pd.DataFrame(array_num, columns=self.num_col)

        # Join
        df = pd.concat([df_num, df_cat], axis=1)
        return df, self.s_scaler

    
x_train, s_scaler = train_preprocessing(cat_col, num_col, x_train, lb_encode, oh_encoder).fit_transform()
x_train.shape
```

#### Find useful features (F Test Regression)

```python
from sklearn.feature_selection import f_regression

f_value, p_value = f_regression(x_train, y_train)
df_p_value = pd.DataFrame(p_value ,index=x_train.columns, columns=['p_value']).sort_values(by=['p_value'], axis=0, ascending=True)

useful_features = df_p_value[df_p_value['p_value'] < 0.1].index
print('Number of useful features: ', len(useful_features))
useful_features
```

```python
x_train = x_train[useful_features]
x_train.head()
```

#### Model

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
y_pred = lin_reg.predict(x_train)

abs_coef = abs(lin_reg.coef_)
coef = lin_reg.coef_
pd.DataFrame({'coef': coef,
              'abs_coef': abs_coef
              }, index=lin_reg.feature_names_in_).sort_values(by=['abs_coef'], ascending=False)
```

```python
coef_of_determination = lin_reg.score(x_train, y_train)
coef_of_determination
```
