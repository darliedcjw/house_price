# Preparation of LabelEncoder and OneHotEncoder

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder

y = df.iloc[:, -1].values
x = df.iloc[:, :-1]

cat_col = [col for col in x.columns if df[col].dtype == 'O']

class get_encoder():
    def __init__(self, cat_col, df):
        self.cat_col = cat_col
        self.df = df

    def fit_transform(self):

        lb_encode = []
        lb_encoder = LabelEncoder()
        oh_encoder = OneHotEncoder(sparse=False)

        df_cat = self.df[self.cat_col]

        for col in self.cat_col:
                df_cat[col] = lb_encoder.fit_transform(df_cat[col])
                lb_encode.append(lb_encoder.classes_)

        oh_encoder.fit(df_cat)
            
        return lb_encode, oh_encoder
    
lb_encode, oh_encoder = get_encoder(cat_col, x).fit_transform()
```

# Clean up missing numerical values

```python
x_train = df_train.iloc[:, :-1]
y_train = df_train.iloc[:, -1].values

cat_col = x_train.dtypes[x_train.dtypes == 'O'].index
num_col = x_train.dtypes[x_train.dtypes != 'O'].index

x_train[num_col].isna().sum()[x_train[num_col].isna().sum() != 0]
x_train['LotFrontage'].fillna(0, inplace=True)
x_train['MasVnrArea'].fillna(0, inplace=True)
x_train['GarageYrBlt'].fillna(0, inplace=True)
```

```python
x_train.head()
```

```python
y_train
```

```python
from sklearn.preprocessing import StandardScaler

class train_preprocessing():
    def __init__(self, cat_col, num_col, df, lb_encode, oh_encoder):
        self.cat_col = cat_col
        self.num_col = num_col
        self.df = df
        self.lb_encode = lb_encode
        self.lb_encoder = LabelEncoder()
        self.s_scaler = StandardScaler()
        self.oh_encoder = oh_encoder
        

    def fit_transform(self):

        # Categorical
        df_cat = self.df[self.cat_col]
        for idx, col in enumerate(self.cat_col):
            self.lb_encoder.fit(self.lb_encode[idx])
            df_cat[col] = self.lb_encoder.transform(df_cat[col])
        
        array_cat = self.oh_encoder.transform(df_cat)
        df_cat = pd.DataFrame(array_cat, columns=self.oh_encoder.get_feature_names_out())

        # Numerical
        df_num = self.df[self.num_col]
        array_num = self.s_scaler.fit_transform(df_num)
        df_num = pd.DataFrame(array_num, columns=self.num_col)

        # Join
        df = pd.concat([df_num, df_cat], axis=1)
        return df, self.s_scaler

    
x_train, s_scaler = train_preprocessing(cat_col, num_col, x_train, lb_encode, oh_encoder).fit_transform()
x_train.shape
```

# Feature Selection

```python
df_num = x_train[num_col]
df_cat = x_train.drop(num_col, axis=1,)
```

## Numerical Variables

#### Variance Threshold

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import VarianceThreshold

variance_selector = VarianceThreshold(threshold=0.2)
variance_selector.fit_transform(df_num, y_train)
variance_name = variance_selector.get_feature_names_out()
df_varthres = pd.DataFrame(variance_name, columns=['num_name'])
df_varthres['variance_threshold'] = 1
df_varthres
```

#### F-Regression

```python
from sklearn.feature_selection import f_regression

_, p_value = f_regression(df_num, y_train)
freg_name = pd.Series(p_value, index=df_num.columns)[pd.Series(p_value, index=df_num.columns)<0.01].index
df_freg = pd.DataFrame(freg_name, columns=['num_name'])
df_freg['f_reg'] = 1
df_num_select = pd.merge(df_varthres, df_freg, on='num_name', how='outer')
df_num_select.dropna(inplace=True)
```

## Categorical Variables

```python
variance_selector = VarianceThreshold(threshold=0.2)
variance_selector.fit_transform(df_cat, y_train)
variance_name = variance_selector.get_feature_names_out()
df_varthres = pd.DataFrame(variance_name, columns=['cat_name'])
df_varthres['variance_threshold'] = 1
df_varthres
```

```python
_, p_value = f_regression(df_cat, y_train)
freg_name = pd.Series(p_value, index=df_cat.columns)[pd.Series(p_value, index=df_cat.columns)<0.01].index
df_freg = pd.DataFrame(freg_name, columns=['cat_name'])
df_freg['f_reg'] = 1
df_cat_select = pd.merge(df_varthres, df_freg, on='cat_name', how='outer')
df_cat_select.dropna(inplace=True)
```

```python
x_test = df_test.iloc[:, :]

cat_col = x_test.dtypes[x_test.dtypes == 'O'].index
num_col = x_test.dtypes[x_test.dtypes != 'O'].index

num_ms_col = x_test[num_col].isna().sum()[x_test[num_col].isna().sum() != 0].index

for col in num_ms_col:
    x_test[col].fillna(0, inplace=True)

class test_preprocessing():
    def __init__(self, cat_col, num_col, df, s_scaler, lb_encode, oh_encoder):
        self.cat_col = cat_col
        self.num_col = num_col
        self.df = df
        self.lb_encode = lb_encode
        self.lb_encoder = LabelEncoder()
        self.s_scaler = s_scaler
        self.oh_encoder = oh_encoder
        

    def fit_transform(self):

        # Categorical
        df_cat = self.df[self.cat_col]
        for idx, col in enumerate(self.cat_col):
            self.lb_encoder.fit(self.lb_encode[idx])
            df_cat[col] = self.lb_encoder.transform(df_cat[col])
        
        array_cat = self.oh_encoder.transform(df_cat)
        df_cat = pd.DataFrame(array_cat, columns=self.oh_encoder.get_feature_names_out())

        # Numerical
        df_num = self.df[self.num_col]
        array_num = self.s_scaler.transform(df_num)
        df_num = pd.DataFrame(array_num, columns=self.num_col)

        # Join
        df = pd.concat([df_num, df_cat], axis=1)
        return df

x_test = test_preprocessing(cat_col, num_col, x_test, s_scaler, lb_encode, oh_encoder).fit_transform()
x_test = x_test[col_train]
x_test.shape
```
